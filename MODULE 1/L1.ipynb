```python
def fullpyramid(n):
      for i in range(1,n+1):
        for j in range(n-i):
            print(" " , end="")
      
        for k in range(1,2*i):
           print("*", end="")
      print()     
      
    
def invpyramid(n):
  for i in range(n,0,-1):
    for j in range(n-i):
      print(" " , end="")

    for k in range(2*i -1):
      print("*", end ="")
    print()

fullpyramid(5)
invpyramid(3)

    
    
```

```python
#FACTORIAL OF A NUMBER

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1) 
num = int(input("Enter a number to find its factorial: "))
result = factorial(num)
print(f"The factorial of {num} is: {result}")
```

```python
#BODY MASS INDEX (BMI) CALCULATOR

def calculate_bmi(w,h):
    bmi = w / (h ** 2)
    return bmi
w = float(input("Enter your weight in kg: "))
h = float(input("Enter yourt height in metres:"))
bmi=calculate_bmi(w,h)
print(f"Your BMI is: {bmi}")

if bmi<18.5:
    print("Underweight")
elif bmi>=18.5 and bmi<24.9:
    print("Healthy weight")
elif(bmi>=25 and bmi <29.9):
    print("Overweigtht")
else :
    print("You are fucking Obese")
```

```python
#   CALCULATOR

choice=int(input("Enter 1 for addition, 2 for subtraction, 3 for multiplication, 4 for division: "))
if choice not in [1,2,3,4] :
    print("Invalid choice. Please enter a number between 1 and 4.")
else :
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    
    match choice:
        case 1:
            result=num1 + num2
            print(f"The sum of {num1} nd {num2} is : {result}")
        case 2:
            result=num1 - num2
            print(f"The difference of {num1} and {num2} is : {result}") 
        case 3:
            result=num1 * num2              
            print(f"The product of {num1} and {num2} is : {result}")
        case 4:
            if num2 != 0:
                result=num1 / num2
                print(f"The quotient of {num1} and {num2} is : {result}")
            else:
                print("Error: Division by zero is not allowed.")
        case _:
            print("Invalid choice. Please enter a number between 1 and 4.")
```

```python
#PRINT THE ELEMENT IN ROW WISE FASHION (SNAKE PATTERN)

numr = int(input("Enter the number of rows :"))
numc = int(input("Enter the number of columns :"))
for i in range(numr):
    if i%2 == 0:
        for j in range(numc):
            print(f"{i*numc + j + 1} ", end="")
    else:
        for j in range(numc-1,-1,-1):
            print(f"{i*numc + j + 1} ", end="")
    print()
```

```python
#snake pattern in reverse order

r=int(input("Enter the number of rows :"))
c=int(input("Enter the number of columns :"))
for i in range(r-1,-1,-1):
    if i%2 == 0:
        for j in range(c):
            print(f"{i*c + j + 1} ", end="")
    else:
        for j in range(c-1,-1,-1):
            print(f"{i*c + j + 1} ", end="")
    print()
```
