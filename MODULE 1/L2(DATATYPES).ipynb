```python
#LISTS 
 

thislist = ["hello", "this" , "is", "a" , "list"]
print(thislist)

#OPERATIONS ON A LIST : 

#1.ACCESSING ELEMENTS :
#Done based on indexing from zero for first object/variable .
#Negative index shows the index of variable from the end . i.e if index = -2 , shows the seocnd last object/variable 

#examples: returns objects from banana to kiwi (excluding 2nd object)
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:5]) 

#prints the elements from the last element till the start (excluding the last element)
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[:-1])


#2.UPDATING THE ELEMENTS : 

#updates the second and third element
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
thislist[1:3]= ["boyy","mannn"]
print(thislist)

#3.ADDING LIST ITEMS :
# TO ADD AT END USE APPEND() :

thislist = ["apple", "banana", "cherry"]
thislist.append("appended at enddd")
print(thislist)


#TO ADD AT RANDOM DESIRED POSTION USE INSERT() :

thislist = ["apple", "banana", "cherry"]
thislist.insert(2 , "inserted after pos")
print(thislist)

#TO ADD ANOTHER LIST/TUPLE USE EXTEND()

lista = ["a","b","c"]
listb = ["x","y","z"]
lista.extend(listb)
print(lista)


#4.DELETING ELEMENTS IN LIST :

#FOR DELETING BY KEY USE REMOVE()
thislist = ["apple", "banana", "cherry"]
thislist.remove("banana")
print(thislist)

#FOR DELETING USING INDEX USE POP(INDEX)

lista = ["a","b","c"]
lista.pop(2)
print(lista) 

#FOR CLEARING ALL ELEMENTS IN LIST USE - CLEAR() i.e clears all fields but the list remains in the memory
#FOR DELETING NETIRE LIST USE - DEL(index) 
#FOR SORTING LISTS USE :
# Ascending sort : thislist.sort() 
# Descending sort : thislist.sort(reverse = TRUE)

thislist = ["apple", "banana", "cherry"]
thislist.sort(reverse=True)
print(thislist)


#Method	Description
append()	Adds an element at the end of the list
clear()	Removes all the elements from the list
copy()	Returns a copy of the list
count()	Returns the number of elements with the specified value
extend()	Add the elements of a list (or any iterable), to the end of the current list
index()	Returns the index of the first element with the specified value
insert()	Adds an element at the specified position
pop()	Removes the element at the specified position
remove()	Removes the item with the specified value
reverse()	Reverses the order of the list
sort()	Sorts the list
```

```python
#TUPLES 

# Tuple items are ordered, unchangeable, and allow duplicate values.
# Tuple items are indexed, the first item has index [0], the second item has index [1] etc.
# tuples are ordered, it means that the items have a defined order, and that order will not change.
# Tuples are unchangeable, meaning that we cannot change, add or remove items after the tuple has been created.

#CREATING TUPLES

#Direct defintion
singeleletuple = ("apple",) #Note the comma after the element its essential 
print(type(singeleletuple)) 

#Using Tuple constructor 
# thistuple = tuple(("apple", "banana", "cherry")) # note the double round-brackets
# print(thistuple)

#Accessing elements of Tuples 

thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango") # -1 index_element i.e mango is excluded 
print(thistuple[-4:-1])

#Modifying the elements of Tuples

# We know that the tuple is immutable ,but there is a workaround .
# Convert Tuple -> List -> Modify the list 
# or add two tuples 

tuplees = ( "a" , "b" , "c")
tupleintolist = list(tuplees)
tupleintolist.append("added to list(tuple)")
print(tupleintolist)


tuplees = ( "a", "b", "c")
singel_ele_tuple=("d" ,)
tuplees += singel_ele_tuple
print(tuplees)


#Unpacking a Tuple 
fruits = ("apple", "banana", "cherry")
(red, yellow, burgundy) = fruits
print(red)   
print(yellow)
print(burgundy)

#with asterick operator
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")
(green, *yellow, red) = fruits
print(green)  # green is the first element of the tuple
print(yellow) # yellow is a list which contains the elements from the second element to the second last element of the tuple
print(red)   #red is the last element of the tuple

#Methods in tuples
# count()	Returns the number of times a specified value occurs in a tuple
# index()	Searches the tuple for a specified value and returns the position of where it was found 
```

```python
#DICTIONARY 

# Dictionaries are used to store data values in key:value pairs.
# A dictionary is a collection which is ordered*, changeable and do not allow duplicates.
# Note : Dictonaries are ordered as of Python version 3.7 and later. In Python 3.6 and earlier, dictionaries are unordered. 

dict = {"Brand": "Ford" , "Model" : "Mustangs" , "year"  : 1964 , "colors": ["red","blue","white"] }
print(dict)

x= dict.get("Model") # returns the value of the key "Model"
print(x)

all_keys =dict.keys()
print(all_keys)


#Nested Dictionaries

myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },   
    "child2" : {
        "name" : "Tobias",
        "year" : 2007
    },
    "child3" : {
        "name" : "Linus",
        "year" : 2011
    }
}
# print(myfamily)

#print(myfamily["child2"]["name"]) # access the name of child2

for x in myfamily:
    print(x) # prints the keys of the dictionary 

    for y in myfamily[x]:
            print(f"{y}: {myfamily[x][y]}") # prints the key and value of the nested dictionary


#Creating a Dicitionary using fromkeys() method

keys =["key1","key2","key3"]
value="xytsz"
mydict =dict.fromkeys(keys, value)
print(mydict)


#Using Default dict from collections import defaultdict

from collections import defaultdict
counter = defaultdict(int) # creates a default dictionary with default value of int (0)
for letter in "apple":
    counter[letter] += 1  # increments the value of key "apple" by 1
print(counter) # prints defaultdict(<class 'int'>, {'apple': 1})

#ASSIGNMENT AND COPYING 

x = { "brand": "Ford",
  "model": "Mustang",
  "year": 1964}
y1 = x # creates a reference to the dictionary x
y2 = x.copy() # creates a copy of the dictionary x

x is y2 # returns False because x and y1 are two different objects in memory
x is y1 #returns True beacuse x anb y2 are the same object in memory



#Dictionary Methods

# Method	Description
# clear()	Removes all the elements from the dictionary
# copy()	Returns a copy of the dictionary
# fromkeys()	Returns a dictionary with the specified keys and value
# get()	Returns the value of the specified key
# items()	Returns a list containing a tuple for each key value pair
# keys()	Returns a list containing the dictionary's keys
# pop()	Removes the element with the specified key
# popitem()	Removes the last inserted key-value pair
# setdefault()	Returns the value of the specified key. If the key does not exist: insert the key, with the specified value
# update()	Updates the dictionary with the specified key-value pairs
# values()	Returns a list of all the values in the dictionary 




    
```

```python
#DOUBLE DICTIONARY USING COPY METHOD :

import copy


x = { 'a' : {'python' :' 2.7'} ,  'b' : { 'python ' : '3.6'}} 

y = x.copy()

x['a']['python'] = '3.8'
print("This is the original dictionary:" , x) # prints the original dictionary with the updated value
print("This is the copied dictionary:" , y) # prints the copied dictionary with the updated value of 'python' key in 'a' key because copy() method creates a shallow copy of the dictionary and does not create a new object in memory for the nested dictionary

#DEEP COPY 
x = { 'a' : {'python' :' 2.7'} ,  'b' : { 'python ' : '3.6'}}
y = copy.deepcopy(x)
print(y)

x['a']['python'] = '3.8'
print("This is the original dictionary:" , x) # prints the original dictionary with the updated value
print("This is the copied dictionary:" , y) # prints the copied dictionary with the original value of 'python' key in 'a' key
```

```python
#SHORTEST STRING IN A LIST 

list =[ "abc " , "bcd" , "bcdefg" , "ab" , "cddc" , "opq" ]

def shortest_string(list):
    key = list[0]
    for x in list:
        if len(x) <= len(key):
            key = x 
    print("the shortest string is:" , key)

shortest_string(list)
```

```python
#SHORTEST AND LONGES STRING USING SORT 

list =[ "abc " , "bcd" , "bcdefg" , "ab" , "cddc" , "opq" ]
list.sort(key =len)
print("The shortest string is :" , list[0])
print("The longest string is :" , list [-1])
```

```python
#DICTIONARY FOR STUDENT DETAILS USING INPUT AND DISPLAY ITS  DETAILS 

student_details={}
n=input("Enter the number of students:")
for i in range (int(n)):
    usn = int(input("Enter the USN:"))
    name = input("Enter the name :")
    semester = int(input("Enter the semester:"))
    cgpa =float(input("Enter the CGPA:"))
    student_details[usn] = {
        "USN": usn ,
        "name": name,
        "cgpa": cgpa,
        "sem": semester ,
    } 

    print(student_details) 

#Searching for student by USN

search_usn =int(input("Enter the USN to search :"))
if search_usn in student_details:
    print(f"Details of student with USN {search_usn}:")
    print(student_details[search_usn])
else:
    print(f"Student with USN {search_usn} not found.")


    
```
