```python
#RECURSIVE FUCNTION CALL :

# A function that calls itself , is called recursive function 

def factorial(n):
    # n = int(input("Enter a number : "))
    if ( n == 0 or n ==1 ):
        return 1
    else: 
        return n*factorial(n-1)
    
print(factorial(8))

```

```python
#FIBONACCI SERIES :

def fibonacci(n):
    n = int(input("Enter a number :"))
    if(n ==0):
        return 0
    elif(n==1):
        return 1 
    else :
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
#Product of two sets 

def product(set1 , set2):
    result = set()
    for i in set1:
        for j in set2:
            result.add((i,j))
    return result 

set1 = { 1, 2 ,3 , 4, 5 ,6 }
set2 = { 4, 5, 6 , 5, 6,  7}
result = product(set1, set2)
print(result)
```

```python
#All possible combination of rolling a dice thrice :

def comb(set1 , set2 , set3):
    result = set()
    for i in set1:
        for j in set2:
            for k in set3:
                result.add((i,j,k))
    return result

set1 = { 1, 2 ,3 , 4, 5 ,6 }
set2 = { 1, 2 ,3 , 4, 5 ,6 }
set3 = { 1, 2 ,3 , 4, 5 ,6 }

result= comb(set1, set2 , set3)
print(result)
```

```python
#TOWER OF HANOI - Takes (2^n - 1) moves to move n disk using three rods 


def toh(n , source , aux , target):
    if(n==1):
        print("Move disk 1 from {} to {}".format(source, target))
        return 
    toh(n-1 , source , target, aux)
    print("Move disk {} from {} to {}".format(n , source , target)) #here the auxilary rod acts as the dest rod , look ovr the ordr of arguymnets that is getting called in the function
    toh(n-1 , aux , target , source)
n = int(input("Enter the number of disks : "))
toh(n , 'A' , 'B' , 'C')    
    
```

```python
#FIBONNACCI SERIES 

def fib(n):
    if n <= 1 :
        return 1 
    else :
        return fib(n-1)+fib(n-2)
    
result=fib(10)
print(result)
```

```python
#Sum of n natural numbers using recursion 

def sum(n):
    # n = int(input("Enter a number :"))
    if n == 0 :
        result = 0
    else:
        result = n + sum(n-1)
    return result 

n = int(input("Enter a number : "))
sum(n)
```

```python
#FIBONNACI USING BINET'S FORMULA 


def binets_fib(n):
    phi = (1 + 5 ** 0.5) / 2
    psi = (1 - 5 ** 0.5) / 2
    fib_n = (phi ** n - psi ** n) / (phi - psi)
    return round(fib_n)


n = int(input("Enter a number : "))
binets_fib(n)
```
