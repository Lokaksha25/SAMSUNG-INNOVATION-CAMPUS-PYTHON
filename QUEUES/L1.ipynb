```python
QUEUE :

FIRST IN FIRST OUT (FIFO)

We can implement queue using list in python. We can use append() to add item to the end of the list and pop(0) to remove item from the front of the list.
```

```python
#IMPLEMENTATION OF QUEUE IN PYTHON:

class queue:

    def __init__(self):
        self.queue=[]

    def enqueue(self,item):
        self.queue.append(item)
        print(f"Added {item} to the queue")

    def dequeue(self):
        if len(self.queue)== 0:
            return "queue is empty" 
        else:
            print(f"Item {self.queue.pop(0)} removed from queue")
    
    def peek(self):
        if len(self.queue)==0:
            return "queue is empty"
        else:
            return self.queue[0]
        
    def print_queue(self):
        if len(self.queue)==0:
            return "queue is empty"
        else:
            print("Queue:",self.queue)
        
q=queue()
q.enqueue(100)
q.enqueue(200)
q.enqueue(300)
q.enqueue(400)
print(q.print_queue())
print(q.dequeue())
print(q.print_queue())
print(q.peek())
```

```python
#JOSEPHUS PROBLEM:

#QUEUE IMPLEMENTATION OF JOSEPHUS PROBLEM:

def josephus(n,k):
    queue= []
    for i in range(1, n+1):
        queue.append(i)
    while len(queue)>1:
        for i in range(k-1):
            queue.append(queue.pop(0))
        queue.pop(0)
    return queue[0]

n = int(input("Enter the number of people in the circle: "))
k = int(input("Enter the skipping step size: "))
print(f"The survivor is {josephus(n, k)}")
```
