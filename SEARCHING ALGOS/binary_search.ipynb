```python
#EGG DROP PROBLEM USING BINARY SEARCH :

from random import randint

height = int (input("Input the numbr of floors :"))
breaking = randint(1 , height)

def do_experiment ( floor , breaking):
    return floor >= breaking

def find_highest_safe_floor(height , breaking):
    low , high  = 1 , height
    while low < high:
        mid= (low+high)//2
        if do_experiment(mid , breaking):
            high = mid
        else:
            low = mid + 1
    return low 

floor = find_highest_safe_floor(height, breaking)
print(f"The egg will break at {floor}th floor")
```

```python
# BINARY SEARCH USING ITERATION :

def binsearch(nums , key ):
    low ,high = 0 , len(nums)-1 
    while low <= high:
        mid = low -( high + low )//2
        if nums[mid] == key:
            print(f"The element : {key} is found at index {mid}")
        elif nums[mid] >key:
            high = mid -1 
        else:
            low = mid + 1
    print("The element is not found !! ")

arrey = [ 10 , 20 , 30 , 40 , 78 , 89 , 91]

binsearch(arrey , 67)
```

```python
# BINARY SEARCH WITH DIVIDE AND CONQUER :


def binsearch(nums , x, low , high):
    if low > high:
        print("The element is not found !!") 
    else:
        mid = (low + high)//2
        if nums[mid] == x :
            print(f"The element : {x} is found at index : {mid}") 
        elif nums[mid]>x:
            return binsearch(nums , x , low , mid-1)
        else:
            return binsearch(nums , x , mid + 1  , high)
        
arrey = [ 10 , 20 , 30 , 40 , 78 , 89 , 91]

binsearch(arrey , 78 , 0 , len(arrey))
```
