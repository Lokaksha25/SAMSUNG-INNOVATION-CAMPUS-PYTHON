```python
def seqsearch(nums , x):
    for i in range(len(nums)):
        if  x == nums[i]:
            print(f"The element {x} is found at {i}")
            break
    else :
        print("not found")
        
        

n = [ 1, 2 , 3, 6, 8 ,11]

seqsearch(n , 6)
```

```python
#FINDING THE LARGEST ELEMENT 

nums = [12 , 34 , 52 , 55 , 67 , 89 , 91  , 19]
n=len(nums)


for i in range(n):
    max = nums[0]
    if nums[i] > max:
        max = nums[i]
print(f"The largest element is : {max}")

```

```python
#TO FIND THE NO OF TIMES A SPECIFIC WORD WAS ENTERED BY USER 

s = input("Enter a sentence :").split()

x = input("input a word to search :")

count = 0

for i in s :
    if i == x:
          count += 1
print(f"The word {x} appeared {count} times")
```
