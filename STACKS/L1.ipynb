```python
#IMPLEMENTATION OF STACKS IN PYTHON:

class stack:

    def __init__(self):
        self.stack=[]
    
    def push(self,item):
        self.stack.append(item)
        print(f"Item {item} added to stack")
    
    def pop(self):
        if len(self.stack) == 0:
            return "stack is empty"
        else:
            print(f"Item {self.stack.pop()} removed from stack")

    def seek(self):
        if len(self.stack)==0:
            return "stack is empty"
        else :
                return self.stack[-1]
        
    def isempty(self):
        return len(self.stack)==0
    

s = stack()
s.push(100)
s.push(200)
s.push(300) 
s.push(400)
print(s.pop())
print(s.seek())

```

```python
#VALIDATION OF BRACKETS USING STACKS :


def isValid(s):
    stack =[]
    opening = "{[("
    closing = "}])"

    for char in s:
        if char in opening :
            stack.append(char)
        elif char in closing:
            if len(stack)==0:
                return False
            if opening.index(stack.pop()) != closing.index(char):
                return False

    return len(stack)==0    

str = input("Input the string of parantheses")

if isValid(str) == True:
    print(f"The expression {str} is Valid")
     
else:
    print(f"The expression {str} is Invalid")



```

```python
#Reversing of string using Stacks 

def rev(s):
    stack = []
    rev = []

    for char in s :
        stack.append(char)
    
    while(len(stack)!= 0):
        rev.append(stack.pop())

    return "".join(rev)
       

exp = input(print("Enter the string :"))

print(f"Before reversing : {exp}")

print("After reversing : ", (rev(exp)))

```

```python
#LONGEST VALID PARENTHESES

#Question : Given a string containing just the characters '(', ')', '{', '}', '[' and ']', find the length of the longest valid (well-formed) parentheses substring.


def longestValidParentheses(s):
    stack = [-1]
    max_length = 0

    for i, char in enumerate(s):
        if char in "([{":
            stack.append(i)
        else:
            stack.pop()
            if len(stack)==0:
                stack.append(i)
            else:
                max_length = max(max_length , i -stack[-1])
                return max_length   
            
s = input("Enter the string of parantheses :")
print(f"The length of longest valid parantheses is : {longestValidParentheses(s)}")
```

```python
#TRAPPING RAIN WATER 

#Question : Given n non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it can trap after raining.

def trap(height):
    if not height:
        return 0
    
    left , right = 0 
    left_max , right_max = height[left] , height[right]
    water_trapped = 0

    while left <right:
        if left_max < right_max:
            left +=1
            left_max = max(left_max , height[left])
            water_trapped += max(0 , left_max - height[left])   
        else:
            right -=1
            right_max = max(right_max , height[right])
            water_trapped += max(0 , right_max - height[right])
    return water_trapped

height = list(map(int , input("Enter the elevation map :").split()))
print(f"The amount of water trapped is : {trap(height)}")
```
