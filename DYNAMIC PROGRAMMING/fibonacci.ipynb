```python
#RECURSIVE FIBONACCI SERIES:

def fibonacci(n):
    if n == 0 or n ==1:
        return n
    else:
        return fibonacci(n-1)+fibonacci(n-2)
    

num = int(input("Enter the number :"))

fibonacci(num)
```

```python
#  DYNAMIC PROGRAMMING APPROACH :

F = { 0 : 0 , 1 : 1 }

def fib2(n):
    if n in F :
        return F[n]
    else:
        F[n] = fib2(n-1) + fib2(n-2)
        return F[n]

num = int(input("Enter the number :"))

fib2(num)
```

```python
#PASCALS TRIANGLE USING RECURSION :


def bin1(n,k):
    if k ==0 or k ==n:
        return 1 
    else:
        return bin1( n - 1 , k - 1 ) + bin1( n - 1 , k )
    
num1 = int(input("Enter the value of 'n' : "))
num2= int (input("Enter the value of 'k' : "))

bin1(num1,num2)
```

```python
#PASCALS TRIANGLE USING DYNAMIC PROGRAMMING :


def bin2(n,k):
    B = [[0]*(i+1) for i in range (n+1)]
    
    for i in range(n+1):
        for j in range(i+1):
            if j ==0 or j == i :
                B[i][j] = 1
            else:
                B[i][j] = B[i-1][j-1] + B [i-1][j]
    return B[n][k]

num1 = int(input("Enter the value of 'n' : "))
num2= int (input("Enter the value of 'k' : "))

print(f"The binomial of ( {num1} , {num2} ) is : { bin1(num1,num2)} ")
```

```python
#MAXIMUM SUB ARRAY :

def max_sub_array(arr):
    if not arr:
        return 0
    
    
    max_so_far = arr[0]
    current_max = arr[0]
    
    for i in range(1, len(arr)):
        
        current_max = max(arr[i], current_max + arr[i])
        
       
        if current_max > max_so_far:
            max_so_far = current_max
            
    return max_so_far

S = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print(f"The maximum subarray sum is: {max_sub_array(S)}") 
```

```python
#MAXIMUM SUB ARRAY BY DP USING KANDANE'S ALGORITHM :

def max_sub_array_dp(arr):
    n = len(arr)
    if n == 0: return [], 0
    
  
    dp = [0] * n
    starts = [0] * n
    
    dp[0] = arr[0]
    starts[0] = 0
    
    max_sum = dp[0]
    best_end = 0

    for i in range(1, n):
        if arr[i] > arr[i] + dp[i-1]:
            dp[i] = arr[i]
            starts[i] = i # New start
        else:
            dp[i] = arr[i] + dp[i-1]
            starts[i] = starts[i-1] 
            
        if dp[i] > max_sum:
            max_sum = dp[i]
            best_end = i

    # Retrieve the subarray
    best_start = starts[best_end]
    sub_array = arr[best_start : best_end + 1]
    
    return sub_array, max_sum

S = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
sub, total = max_sub_array_dp(S)
print(f"The maximum Subarray is : {sub} |  Sum: {total}")
```
