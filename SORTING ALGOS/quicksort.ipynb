```python
# PARTITION AND MERGING USIGN LOMUTO PARTITION SCHEME :

def partition_lomuto(arr, low, high, descending, swaps):
    # Lomuto always chooses the last element as the pivot
    pivot = arr[high]
    
    # 'i' tracks the boundary of where the smaller/larger elements end
    i = low - 1
    
    for j in range(low, high):
        # Determine sorting direction
        if descending:
            # We want numbers LARGER than the pivot on the left
            condition = arr[j] > pivot
        else:
            # We want numbers SMALLER than the pivot on the left
            condition = arr[j] < pivot
            
        if condition:
            i += 1
            # Swap the current element into the correct boundary zone
            arr[i], arr[j] = arr[j], arr[i]
            
            # Only count it as a swap if it didn't just swap with itself
            if i != j: 
                swaps[0] += 1
                
    # Finally, put the pivot exactly where it belongs (right after 'i')
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    
    if (i + 1) != high:
        swaps[0] += 1
        
    return i + 1
```

```python
#PARTITION USING HOARSE PARTITION SCHEME:

def quicksort(arr):
    if len(arr) <= 1 :
        return arr
    
    smaller , equal , larger = [] , [] , []

    pivot = arr[-1]

    for num in arr :
        if num < pivot:
            smaller.append(num)
        elif num == pivot:
            equal.append(num)
        elif num > pivot:
            larger.append(num)

    return quicksort(smaller)+equal+quicksort(larger)
 

arrey = [ 21 , 2 , 3 , 5 , 67 , 243 , 15]

quicksort(arrey)
```
