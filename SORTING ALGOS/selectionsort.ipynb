```python
def selection_sort(nums, descending=False):
    n = len(nums)
    total_swaps = 0

    for i in range(n - 1):
        # Assume the first unsorted element is our target (min or max)
        target_index = i
        
        # Scan the rest of the unsorted list
        for j in range(i + 1, n):
            if descending:
                # If descending, look for the LARGEST remaining number
                if nums[j] > nums[target_index]:
                    target_index = j
            else:
                # If ascending, look for the SMALLEST remaining number
                if nums[j] < nums[target_index]:
                    target_index = j
        
        # If a new min/max was found, swap it into the current position 'i'
        if target_index != i:
            nums[i], nums[target_index] = nums[target_index], nums[i]
            total_swaps += 1
            
    return total_swaps

# --- User Input Section ---
print("What would you like to sort?")
print("1. Numbers")
print("2. Words / Letters (Alphabetical)")
data_type = input("Enter 1 or 2: ")

user_input = input("\nEnter your items separated by spaces: ")

if data_type == '1':
    try:
        x = [float(num) for num in user_input.split()]
    except ValueError:
        print("Error: You selected 'Numbers' but entered text. Exiting.")
        exit()
else:
    x = user_input.split()

order_choice = input("Sort descending (Z-A or Highest-Lowest)? (y/n): ").lower()
is_descending = (order_choice == 'y')

print(f"\n--- Results ---")
print(f"Original list: {x}")

# Call the function and save the returned swap count
swaps_made = selection_sort(x, descending=is_descending)

print(f"Sorted list:   {x}")
print(f"Total swaps:   {swaps_made}")
```
