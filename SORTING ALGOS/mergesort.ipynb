```python
# MERGE SORTING :


#USES TWO FUNCTIONS TO DIVIDE AND CONQUER I.E 
#FOR DIVIDE ITS MERGESORT AND TO CONQUER(COMBINE) WE USE MERGE()
```

```python
#MERGE FUCNTION 

def merge(arr , l , m , r):
    n1 = m - l + 1
    n2 = r - m

    L = [0]*n1
    R = [0]*n2

    for i in range(n1):
        L[i]=arr[l+i]
    
    for j in range(n2):
        R[j]=arr[m+1+j]

    i = j = 0
    k = l

    while i < n1 and j < n2:
        if L[i]<R[j]:
            arr[k]=L[i]
            i += 1
        else:
            arr[k]=R[j]
            j += 1
        k += 1

    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1
    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1
```

```python
def merge(arr , l , m , r):
    n1 = m - l + 1
    n2 = r - m

    L = [0]*n1
    R = [0]*n2

    for i in range(n1):
        L[i]=arr[l+i]
    
    for j in range(n2):
        R[j]=arr[m+1+j]

    i = j = 0
    k = l

    while i < n1 and j < n2:
        if L[i]<R[j]:
            arr[k]=L[i]
            i += 1
        else:
            arr[k]=R[j]
            j += 1
        k += 1

    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1
    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1


def mergesort(arr, l , r):
    if l < r:
         m = l + ( r - l )//2
         mergesort(arr, l , m)
         mergesort(arr, m+1 , r)
         merge(arr , l , m ,r)



arr = [ 10 , 20 , 40 , 34, 12 , 89 , 17]

print(f"Before sorting : {arr}")


mergesort(arr, 0, len(arr) -1 )

print(f"After sorting : {arr}")


   
```
