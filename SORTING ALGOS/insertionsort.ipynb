```python
def insertionsort(arr):
    n = len(arr)

    if n <= 1:
        return 
    for i in range (1, n):
        key= arr[i]
        j = i- 1

        while j >=0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key 

nums = (10 , 45 , 2 , 0 , 0 , 1 , 4 , 67)
print(nums)
insertionsort(nums)
print(nums)
```
