```python
#BUBBLE SORTING ALGORITHM :


def bubblesort(nums):
    n = ( len(nums) -1 )

    for i in range(n):
        swapped = False 
        swaps  = 0 
        for j in range (n-1-i):
            if nums[j] > nums[j+1]:
                nums[j],nums[j+1] = nums[j+1] ,nums[j]
                swapped = True
                swaps += 1 
            
        if swapped == False:
            break
    
x = list(input("Enter the list to the sorted : "))

bubblesort(x)

print(x)




```

```python
def bubblesort(nums, descending=False):
    n = len(nums)
    total_swaps = 0 

    for i in range(n):
        swapped = False 
        
        for j in range(n - 1 - i):
            # Feature 2: Check order direction (Ascending vs Descending)
            if descending:
                # For descending, swap if the current is LESS than the next
                out_of_order = nums[j] < nums[j+1]
            else:
                # For ascending, swap if the current is GREATER than the next
                out_of_order = nums[j] > nums[j+1]

            if out_of_order:
                nums[j], nums[j+1] = nums[j+1], nums[j]
                swapped = True 
                total_swaps += 1  # Increment counter on every successful swap
                
        if not swapped:
            break
            
    return total_swaps # Return the final count


print("What would you like to sort?")
print("1. Numbers")
print("2. Words / Letters (Alphabetical)")
data_type = input("Enter 1 or 2: ")

user_input = input("\nEnter your items separated by spaces: ")


if data_type == '1':

    try:
        x = [float(num) for num in user_input.split()]
    except ValueError:
        print("Error: You selected 'Numbers' but entered text. Exiting.")
        exit()
else:
    
    x = user_input.split()

order_choice = input("Sort descending (Z-A or Highest-Lowest)? (y/n): ").lower()
is_descending = (order_choice == 'y')

print(f"\n--- Results ---")
print(f"Original list: {x}")

swaps_made = bubblesort(x, descending=is_descending)

print(f"Sorted list:   {x}")
print(f"Total swaps:   {swaps_made}")
```
