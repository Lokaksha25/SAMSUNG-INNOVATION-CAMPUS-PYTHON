```python
#GREEDY ALGORITHM 


# Greedy algorithms aim to make the optimal choice at that given moment.
# Each step it chooses the optimal choice, without knowing the future. 
# It attempts to find the globally optimal way to solve the entire problem using this method.

# Note :
# They are only concerned with the optimal solution locally. 
# This means that the overall optimal solution may differ from the solution the algorithm chooses.
```

```python
# Sometimes, Greedy algorithms give the global optimal solution every time. Some of these algorithms are:

# Dijkstra’s Algorithm
# Kruskal’s algorithm
# Prim’s algorithm
# Huffman trees
```
