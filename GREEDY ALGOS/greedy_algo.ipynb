```python
#GREEDY ALGORITHM 


# Greedy algorithms aim to make the optimal choice at that given moment.
# Each step it chooses the optimal choice, without knowing the future. 
# It attempts to find the globally optimal way to solve the entire problem using this method.

# Note :
# They are only concerned with the optimal solution locally. 
# This means that the overall optimal solution may differ from the solution the algorithm chooses.
```

```python
# Sometimes, Greedy algorithms give the global optimal solution every time. Some of these algorithms are:

# Dijkstra’s Algorithm
# Kruskal’s algorithm
# Prim’s algorithm
# Huffman trees
```

```python
 #ACTIVITY SELECTION CODE :


starting = [ 1 , 3, 2, 0 , 5, 8 , 5]
ending = [2 , 4 , 5, 6, 6, 9 , 9 ]


def activity_selection(starting , ending  ):
    result = []
    i = 0
    result.append(i)
    for j in range(1, len(starting)):
        if ending[i] <= starting[j]:
            result.append(j)
            i = j 
        return result 
    

meetings = activity_selection(starting , ending)

maximum = len(meetings)
print(meetings , maximum)
```

```python
# MINIMUM TRIANGLE PATH


triangle = [ 
      [2] ,
    [3 , 4] , 
 [ 0 , 5 , 7],
[5 , 9 , 1 , 3]
]


def find_minimum(row , col , triangle ):
    if row == len(triangle):
        return 0 
    else:
        minimum = min(find_minimum(row +1 , col , triangle), 
                      find_minimum(row+1 , col+1 , triangle ))
        return  (triangle[row][col] + minimum)

minimum = find_minimum(0 , 0 , triangle)
print("The minimum cost of is " , minimum)
```

```python
#MINIMUM TRAINGLE PATH USING DYNAMIC PROGRAMMING :


def solve_triangle_path(triangle):
    n = len(triangle)
    # Create a DP table to store minimum sums at each node
    # We copy the triangle to avoid mutating the original data
    dp = [row[:] for row in triangle]
    # Path table to store the index of the chosen next step
    paths = [[0] * len(row) for row in triangle]

    # Process from the second-to-last row up to the top
    for row in range(n - 2, -1, -1):
        for col in range(len(triangle[row])):
            # Compare the two options below
            left_child = dp[row + 1][col]
            right_child = dp[row + 1][col + 1]
            
            if left_child < right_child:
                dp[row][col] += left_child
                paths[row][col] = col  # Move stayed in the same column
            else:
                dp[row][col] += right_child
                paths[row][col] = col + 1 # Move shifted right

    # Trace the path back from the top
    actual_path = []
    current_col = 0
    for row in range(n):
        actual_path.append(triangle[row][current_col])
        if row < n - 1:
            current_col = paths[row][current_col]

    return dp[0][0], actual_path

# Example Data
triangle = [ 
        [2],
      [3 , 4], 
    [0 , 5 , 7],
   [5 , 9 , 1 , 3]
]

min_cost, path_taken = solve_triangle_path(triangle)

print(f"The minimum cost is: {min_cost}")
print(f"The path traced is: {' -> '.join(map(str, path_taken))}")
```
